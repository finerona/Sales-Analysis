```python
import pandas as pd
```

```python
df=pd.read_csv('raw.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.columns
```

```python
newdf = df[['Order Channel', 'Quantity', 'Unit Price', 'Customer Type', 'Category Group', 'shippingStatus', 'Category', 'Delivery Date',]]
```

```python
newdf.head()
```

```python
 newdf.isna().sum()
```

```python
newdf.head()
```

```python
newdf['Delivery Date'] = pd.to_datetime(newdf['Delivery Date'])
```

```python
newdf.head()
```

```python
newdf['Unit Price'] = newdf['Unit Price'].str.replace(',','').astype(float)
```

```python
newdf.head()
```

```python
newdf['Sales'] = newdf['Quantity'] * newdf['Unit Price']
```

```python
newdf.head()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.hist(newdf['Quantity'], bins = 5)
plt.title('order Quantity Distribution')
plt.xlabel('Quantity')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.hist(newdf['Order Channel'], bins = 5)
plt.title('order Channel')
plt.xlabel('Channel')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.hist(newdf['shippingStatus'], bins = 5)
plt.title('shipping Status')
plt.xlabel('shipping')
plt.ylabel('Frequency')
plt.show()
```

```python
newdf.to_csv('Cleanedshopdata.csv', index=False)
```
